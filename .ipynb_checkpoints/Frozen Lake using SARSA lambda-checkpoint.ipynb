```python
import numpy as np
import pandas as pd
import time
import gym
from IPython.display import clear_output
```

```python
N_STATES_V = 5
N_STATES_H = 5
ACTIONS = ['left','right','up','down']
ALPHA = 0.1
Gamma = 0.9
Greed_F = 0.5
Frame_Time = 0.01
Max_Generations = 200
np.random.seed(2)
```

```python
def choose_action(q_table,S,Greed_F):
    q = q_table.iloc[S,:]
    if ((np.random.uniform() > Greed_F) or ((q == 0).all())) and (Frame_Time != 1):
        a = np.random.choice(ACTIONS)
    else:
        a = q.idxmax()
    return a
```

```python
def update_env(S,gen,step):
    env = envh = np.array([['_']*N_STATES_H])
    for i in range(N_STATES_V-1):
        env = np.vstack((env,envh))
    env [ N_STATES_V-1 , N_STATES_H-1 ] = 'X'
    if (Frame_Time == 1):
        clear_output()
    print ('Generation : ' + str(gen+1) + '\t Step ' + str(step),end = '\r')
    if S=='Treasure':
        Gen_Sum = 'Generation : ' + str(gen) + '\t Completed in ' + str(step) + ' steps'
        print (Gen_Sum)
    else:
        env[int ((S)/N_STATES_H),(S)%N_STATES_H] = 'O'
        if (Frame_Time == 1):
            print(env)
        time.sleep(Frame_Time)
```

```python
def feedback_from_env(S,A):
    if A == 'right':
        S_ = S + 1
        R = 0
    elif A == 'down':
        S_ = S + 5
        R = 0
    elif A == 'left':
        S_ = S - 1
        R = 0
    elif A == 'up':
        S_ = S - 5
        R = 0
    if S_ < 0:
        S_ = 0
    elif S_ > 23:
        S_ = 'Treasure'
        R = 1
    return S_,R
```

```python
def create_q_table(states):
    q_table = pd.DataFrame(
    np.zeros((states,len(ACTIONS))),
    columns = ACTIONS
    )
    return q_table
```

```python
def simulate(q_table):
    for gen in range(Max_Generations):
        step = 0
        S = 19
        update_env(S,gen,step)
        R = 0
        Terminated = False
        while not Terminated:
            step += 1
            A = choose_action(q_table,S)
            S_,R = feedback_from_env(S,A)
            update_env(S_,gen,step)
            predict_q = q_table.loc[S,A]
            if (S_ != 'Treasure'):
                target_q = R + Gamma*q_table.iloc[S_,:].max()
            else:
                target_q = R
                Terminated = True
            q_table.loc[S,A] += ALPHA * (target_q - predict_q)
            S = S_
    return q_table
```

```python
np.random.seed(2)
Frame_Time = 0.01
Max_Generations = 200
q_table = create_q_table(N_STATES_H*N_STATES_V)
q_table = simulate(q_table)
print('\r\nQ-table:\n',q_table)
```
